```python
response = requests.get("https://data.cityofnewyork.us/resource/8h9b-rp9u.json")
```

```python
arrestData = response.json()
columns = list(arrestData[0])
byBorough = {'Q': {'total': 0, 'female' : 0, 'male': 0}, 'K':{'total': 0, 'female' : 0, 'male': 0}, 'B': {'total': 0, 'female' : 0, 'male': 0}, 'S':{'total': 0, 'female' : 0, 'male': 0}, 'M':{'total': 0, 'female' : 0, 'male': 0}}
ageData = {}
race = {}
ageGenderData = {'25-44':{'F':0, 'M':0}, '18-24': {'F':0, 'M':0}, '45-64': {'F':0, 'M':0}, '65+': {'F':0, 'M':0}, '<18': {'F':0, 'M':0}}
raceData = {}
lawData = {}
for i in arrestData:
    if '2017' in i['arrest_date']:        
        byBorough[i['arrest_boro']]['total'] = byBorough[i['arrest_boro']]['total'] + 1
        if i['perp_race'] in byBorough[i['arrest_boro']]:
            byBorough[i['arrest_boro']][i['perp_race']] = byBorough[i['arrest_boro']][i['perp_race']] + 1
        else:
            byBorough[i['arrest_boro']][i['perp_race']] = 1
        if i['age_group'] in byBorough[i['arrest_boro']]:
            byBorough[i['arrest_boro']][i['age_group']] = byBorough[i['arrest_boro']][i['age_group']] + 1
        else:
            byBorough[i['arrest_boro']][i['age_group']] = 1
        if i['perp_race'] in race:
            race[i['perp_race']] = race[i['perp_race']] + 1
        else:
             race[i['perp_race']]  = 1
        if(i['perp_sex'] == 'F'):
            byBorough[i['arrest_boro']]['female'] = byBorough[i['arrest_boro']]['female'] + 1
            ageGenderData[i['age_group']]['F'] = ageGenderData[i['age_group']]['F'] + 1
        else:
            byBorough[i['arrest_boro']]['male'] = byBorough[i['arrest_boro']]['male'] + 1
            ageGenderData[i['age_group']]['M'] = ageGenderData[i['age_group']]['M'] + 1
        if i['perp_race'] in raceData:
            raceData[i['perp_race']][i['perp_sex']] = raceData[i['perp_race']][i['perp_sex']] + 1
            if i['age_group'] in raceData[i['perp_race']]:
                raceData[i['perp_race']][i['age_group']] =  raceData[i['perp_race']][i['age_group']] + 1
            else:
                 raceData[i['perp_race']][i['age_group']] = 1
        else:
            raceData[i['perp_race']] = {'F': 0, 'M': 0}
        if i['age_group'] in ageData:
            ageData[i['age_group']] = ageData[i['age_group']] + 1
        else:
            ageData[i['age_group']] = 1
        if i['law_cat_cd'] in lawData:
            lawData[i['law_cat_cd']] = lawData[i['law_cat_cd']] + 1
        else:
            lawData[i['law_cat_cd']] = 1
for raceKey in race:
    for borough in byBorough:
        if raceKey not in byBorough[borough]:
            byBorough[borough][raceKey] = 0
for raceKey in raceData:
    for age in ageData:
        if age not in raceData[raceKey]:
            raceData[raceKey][age] = 0
for age in ageData:
    for borough in byBorough:
        if age not in byBorough[borough]:
            byBorough[borough][age] = 0
```

# Chi-Square Test for Homeginity

Below, I perform the Chi-Square Test for Homeginity across different races, ages, and sexes across the five boroughs. I have 5 outcomes, where each outcome is a borough and 2 distributions. So my degrees of freedom is (5-1)(2-1) = 4.

```python
dist = stats.chi2(df=4)
print('Rejection Region: Greater than', np.round(dist.ppf(0.95), 2))
fig = plt.figure(figsize=(10,10))
x = np.linspace(dist.ppf(0.0001), dist.ppf(0.9999), 100)
plt.plot(x, dist.pdf(x))
plt.plot([dist.ppf(0.95), dist.ppf(0.95)],[0, dist.pdf(dist.ppf(0.95))], linestyle='dotted', color='#1f77b4')
X_fill = np.linspace(dist.ppf(0.95), dist.ppf(0.999))
Y = dist.pdf(X_fill)
plt.fill_between(X_fill, 0, Y, facecolor='red')
plt.ylim(bottom=0)
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density Function')
fig.suptitle('Chi-Square Distribution with 4 Degrees of Freedom', fontsize=15, y=0.92)
```
